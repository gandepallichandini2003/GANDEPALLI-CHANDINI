**REMOVING DUPLICATES**

**Import Pandas**

```python
import pandas as pd
```

**Load the Dataset**

```python
df = pd.read_excel('/hospital data analysis dataset.xlsx')
```

**Display Data**

```python
df.head()
```

**Check Dataset Info**

```python
df.info()
```

Find Duplicate Rows

```python
df.duplicated().sum()
```

View Duplicate Rows

```python
df[df.duplicated()]
```

Remove Duplicate Rows

```python
df_cleaned = df.drop_duplicates()
```

Save Cleaned Data

```python
df_cleaned.to_excel('cleaned_hospital_data.xlsx', index=False)
```
