**DATA CLEANING**

Handle Missing Values

Check missing values

```python
import pandas as pd
df = pd.DataFrame({'A': [1, 2, None], 'B': [4, None, 6]})
df.isnull().sum()
```

Fill missing values

```python
df.fillna(method='ffill', inplace=True)  # forward fill
```

Remove Duplicate Rows

```python
df = df.drop_duplicates()
```

Handle Incorrect Values

```python
import numpy as np
df.replace(['?', 'NA', 'null'], np.nan, inplace=True)
```

**DATA PREPARATION STEPS**

Feature Selection

```python
df = pd.DataFrame({
    'patient_id': [1, 2, 3, 4, 5],
    'age': [30, 45, 22, 58, 37],
    'gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'disease': ['Flu', 'Cold', 'Flu', 'Allergy', 'Cold'],
    'treatment': ['A', 'B', 'A', 'C', 'B'],
    'other_column': [10, 20, 30, 40, 50]
})
df = df[['patient_id', 'age', 'gender', 'disease', 'treatment']]
```

Encode Categorical Data

```python
df['gender'] = df['gender'].map({'Male': 0, 'Female': 1})
```

Final Clean Check

```python
df.info()
df.isnull().sum()
```

Save Cleaned Dataset

```python
df.to_excel('cleaned_hospital_data.xlsx', index=False)
```
