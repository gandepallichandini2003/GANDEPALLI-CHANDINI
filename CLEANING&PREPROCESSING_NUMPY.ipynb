```python
import pandas as pd
import numpy as np
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
file_name = list(uploaded.keys())[0]
df = pd.read_excel(file_name)
display(df.head())
```

```python
print('Missing values in each column:')
print(df.isnull().sum())
```

```python
data = df.astype(str).to_numpy()
```

**DATA CLEANING USING NUMPY**

Check Missing Values

```python
print(data)
```

Replace Missing Values

```python
data[data == 'nan'] = '0'
data[data == 'NA'] = '0'
data[data == 'null'] = '0'
```

Remove Duplicate Rows

```python
data = pd.DataFrame(data).drop_duplicates().to_numpy()
```

Remove Extra Spaces

```python
data = pd.DataFrame(data).map(lambda x: str(x).strip()).to_numpy()
```

**DATA PREPARATION USING NUMPY**

Select Required Columns

```python
data = data[:, [0, 1, 2]]  # select columns by index
```

Encode Categorical Data

```python
data[:, 2][data[:, 2] == 'Male'] = '0'
data[:, 2][data[:, 2] == 'Female'] = '1'
```

Convert Back to DataFrame

```python
df_cleaned = pd.DataFrame(data, columns=df.columns[:data.shape[1]])
```

Save Cleaned Data

```python
df_cleaned.to_excel('final_cleaned_data.xlsx', index=False)
```
